```python
import numpy as np
import tensorflow as tf
import random
```

```python
m = 10_000
n = 1_000
epochs = 5000

x1_train=[]
x2_train=[]
y_train=[]
x1_test=[]
x2_test=[]
x_test = []
y_test=[]
for i in range(m):
  x1_train.append(random.uniform(-10, 10))
  x2_train.append(random.uniform(-10, 10))
  if x1_train[-1] < -5 or x1_train[-1] > 5:
    y_train.append(1)
  else:
    y_train.append(0)
x_train = np.column_stack((x1_train,x2_train))
y_train = np.array(y_train)

for i in range(n):
  x1_test.append(random.uniform(-10, 10))
  x2_test.append(random.uniform(-10, 10))
  if x1_test[-1] < -5 or x1_test[-1] > 5:
    y_test.append(1)
  else:
    y_test.append(0)
x_test = np.column_stack((x1_test,x2_test))
y_test = np.array(y_test)
```

```python
print(x_train.shape)
print(x_test.shape)
print(x_train[0], y_train[0])
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(2,)),# 2,1
  tf.keras.layers.Dense(2, activation='sigmoid'),
  # tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])

# model.compile(optimizer='sgd',
#               loss=loss_fn,
#               metrics=['accuracy'])

# model.compile(optimizer='rmsprop',
#               loss=loss_fn,
#               metrics=['accuracy'])

```

```python
model.fit(x_train, y_train, epochs = epochs, verbose = 2)
```

```python
model.evaluate(x_test,  y_test, verbose=2)
```
